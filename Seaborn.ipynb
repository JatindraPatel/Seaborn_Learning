# **SEABORN** :-


Seaborn is a Python data-visualization library built on top of Matplotlib.


It makes creating beautiful, statistical, and customized graphs much easier and faster.






**‚≠ê Key Features of Seaborn**

High-level interface ‚Üí easier and cleaner than Matplotlib.



Beautiful default themes ‚Üí your graphs look professional automatically.

Supports statistical plots ‚Üí regression plots, distribution plots, heatmaps, etc.

Integrates well with pandas DataFrames.

**üìä Common Plots in Seaborn**

sns.lineplot() ‚Üí line chart

sns.barplot() ‚Üí bar chart

sns.countplot() ‚Üí count/frequency chart

sns.heatmap() ‚Üí correlation or matrix heatmap

sns.distplot() / sns.histplot() ‚Üí distributions

sns.boxplot() ‚Üí box & whisker plot

sns.pairplot() ‚Üí multiple relationships visualization

sns.regplot() ‚Üí regression trend line

**üéØ Why Use Seaborn**

Less code than Matplotlib

Better colors and styling

Great for statistical data analysis

Works perfectly with DataFrames (df)

**Difference Between Seaborn and Matplotlib**


| Feature                   | **Matplotlib**                   | **Seaborn**                                                                   |
| ------------------------- | -------------------------------- | ----------------------------------------------------------------------------- |
| **Level**                 | Low-level                        | High-level                                                                    |
| **Ease of Use**           | More code, manual styling        | Easy, automatic styling                                                       |
| **Default Look**          | Basic, plain                     | Beautiful & modern                                                            |
| **Focus**                 | General plotting                 | Statistical plotting                                                          |
| **Best For**              | Custom plots, full control       | Quick, good-looking plots                                                     |
| **Works With DataFrames** | Not very smooth                  | Very smooth (built for DataFrames)                                            |
| **Plot Types**            | Basic plots (line, bar, scatter) | Advanced statistical plots (heatmap, pairplot, distplot, violinplot, regplot) |
| **Customization**         | Very high but requires more code | High but simpler                                                              |
| **Built On**              | Base library                     | Built on top of Matplotlib                                                    |


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([0, 1, 2, 3, 4, 5])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([0, 1, 2, 3, 4, 5],hist=False)
plt.show()
```

# Relational/ Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8-talk')

sns.relplot(x='Total_bill',y='Tip',data=df)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8-talk')

sns.relplot(x='Total_bill',y='Tip',hue='Smoker',data=df)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8-talk')

sns.relplot(x='Total_bill',y='Tip',hue='Smoker',style='Sex',data=df)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8-talk')

sns.relplot(x='Total_bill',y='Tip',hue='Smoker',style='Sex',size='Day',data=df)
plt.show()
```
